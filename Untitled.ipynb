```python
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder,StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import DataSet
train_data = pd.read_csv('mushroom_train.csv')
test_data = pd.read_csv('mushroom_test.csv')
train_data.head()
```

```python
#Now Gathering Information about Dataset
train_data.info()
```

```python
#We can see that the dataset does not have Null Values
#Now since the 23 Instance has object datatype we have to change it in INT 
#For changing datatype we will use Label Encoder 
#It will label different into INT format
label_encoder = LabelEncoder()
for i in train_data.columns:
    train_data[i] = label_encoder.fit_transform(train_data[i])
for i in test_data.columns:
    test_data[i] = label_encoder.fit_transform(test_data[i])
train_data.head()
```

```python
#Now the DataSet is changed into Desired Format
train_data.describe()
```

```python
# Seperating Labels From DataSet
train_labels = train_data.iloc[:,0] #Selecting Row 1
train_data = train_data.iloc[:,1:25] # Selecting Remaining Rows
train_data.head()
```

```python
#Dividing Training Dataset into Two Parts Training and Validation Dataset
#Divide Dataset into 4:1
validation_data = train_data.iloc[1:1130,:]
validation_labels = train_labels.iloc[1:1130]
train_data = train_data.iloc[1131:,:]
train_labels = train_labels.iloc[1131:]
```

```python
#The Description of Dataset Shows it is not Standardised i.e. Range is not -1 to 1
#Standardised Data helps to make Mean of Data approximately equals to 0
scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.fit_transform(test_data)
validation_data = scaler.fit_transform(validation_data).astype(np.float32)
```

```python
#Changing Labels into One HOT encoded vector
def reformat(train_labels):
    train_labels = (np.arange(2) == train_labels[:,None]).astype(np.float32)
    return train_labels
train_labels = reformat(train_labels)
validation_labels = reformat(validation_labels)
validation_data = validation_data.astype(np.float32)#changing validation datatype into Float32
train_data = train_data.astype(np.float32)#changing train_data datatype into Float32
test_data = test_data.astype(np.float32)#changing test datatype into Float32
train_labels.shape
```

```python
#Graph for Tesnorflow Session
graph = tf.Graph()
batch_size = 128 #Size of input used to train Model in one epoch
with graph.as_default():
    tf_train_data = tf.placeholder(tf.float32, shape=(batch_size,24))
    tf_train_label = tf.placeholder(tf.float32, shape=(batch_size,2))
    tf_validation_data = tf.constant(validation_data)
    tf_validation_labels = tf.constant(validation_labels)
    tf_test = tf.constant(test_data)
    step = 0
    weights = tf.Variable(tf.truncated_normal([24,150]))#Weight of Neural Network multiplied by the features
    hidden_weights = tf.Variable(tf.truncated_normal([150,2]))#Hidden layer weights
    hidden_biases = tf.Variable(tf.zeros(2))#hidden layer biases
    biases = tf.Variable(tf.zeros(150))#Biases added to weights
    logit = tf.matmul(tf.nn.dropout(tf.nn.relu((tf.matmul(tf_train_data,weights) + biases)),0.75),hidden_weights)+hidden_biases
    #The stucture of our model is defined by the logit 
    # Layer 1 = (INPUT*Weight) + Biases
    #Then applying RELU activation Function
    #NOW using dropout with keep_prob = 0.75
    #THEN hidden layer = Dropout(Relu(Layer1))*Hidden weights + Hidden Biases 
    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=tf_train_label, logits=logit))
    #Loss function is made to calculate softmax of output and then calculating loss with ONE HOT encoded Vector
    optimizer = tf.train.GradientDescentOptimizer(0.08).minimize(loss)
    #Optimizer will reduced loss with Gradient Descent having learning rate = 0.08
    train_prediction = tf.nn.softmax(logit)#Converting output of our model into softmax probabilities
    validation_prediction = tf.nn.softmax(tf.matmul(tf.nn.relu((tf.matmul(tf_validation_data,weights) + biases)),hidden_weights)+hidden_biases)
    #Validation prediction will be used to predict our validation dataset
    test_prediction = tf.nn.softmax(tf.matmul(tf.nn.relu((tf.matmul(tf_test,weights) + biases)),hidden_weights)+hidden_biases)
    #Test prediction will be used to predict our Test dataset
```

```python
num_steps = 2500 #Number of EPOCHS
#Accuracy Function used to calculate accurary
def accuracy(predictions, labels):   
    return (100.0*np.sum(np.argmax(predictions,1) == np.argmax(labels,1))
            /predictions.shape[0])
with tf.Session(graph=graph) as session:
    tf.global_variables_initializer().run()
    print("Started")
    for step in range(num_steps):
        offset = (step * batch_size) % (train_labels.shape[0] - batch_size)
        #offset will be random number used to select random data from training dataset
        batch_data = train_data[offset:(offset + batch_size), :]
        #Selecting random dataset to train our model
        batch_labels = train_labels[offset:(offset+batch_size), :]
        #Selecting Label for our training data
        feed_dict = {tf_train_data:batch_data, tf_train_label:batch_labels}#Giving Data and label 
        _, l, predictions = session.run(
        [optimizer, loss, train_prediction], feed_dict = feed_dict)
        if(step % 100 == 0):
            print("Step", step, "Loss", l)
            print('Batch accuracy', accuracy(predictions,batch_labels))#Checking Accuracy of Batch Data
    print("Validation Dataset Accuracy",accuracy(validation_prediction.eval(),validation_labels)) #Evaluating Validation Data
    ans = test_prediction.eval()#Evaluating Test Data
```
